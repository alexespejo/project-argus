```python
class User:
    def __init__(self, name, score=0):
        self.name = name
        self.score = score

    def update_score(self, score):  # update users score
        self.score = score
        
    def get_score(self):  # return the user's score
        return self.score

    def get_name(self):  # return the user's name
        return self.name

    def get_user(self):  # return a dictionary of the user
        return {u"name": self.name, u"score": self.score}


class Game:
    # creates a game with a scoreboard
    def __init__(self, scoreboard=[]):
        self.scoreboard = scoreboard

    def play(play = True):
        return play
        
    def update_score(self, score=None):
        if score != None:  # if there is a newly added score it will be added to the scoreboard other wise it will continue
            print(f"New Score {score.get_name()}")
            self.scoreboard.append(score.get_user())
        x = 0
        placeholder = 0
        while x < len(self.scoreboard):  # sort the scores in greatest to least
            y = x
            while y < len(self.scoreboard):
                if self.scoreboard[x].get("score") < self.scoreboard[y].get("score"):
                    placeholder = self.scoreboard[x]
                    self.scoreboard[x] = self.scoreboard[y]
                    self.scoreboard[y] = placeholder
                y += 1
            x += 1

    def get_highscore(self):  # return the highest score
        self.highscore = self.scoreboard[0]
        return self.highscore

    def print_scores(self):  # returns the scoreboard in a string
        scores = ""
        count = 0
        while count < len(self.scoreboard):
            scores += f"{count + 1}. {self.scoreboard[count].get(u'name')}\n"
            count += 1
        print(scores)


# GAME RUN

# game starts with a preset users
game = Game([User("Tim", 5).get_user(), User("Alex", 18).get_user(),
            User("Erin", 19).get_user(), User("Kai", 2).get_user()])

# adds a user named andrew with a score of 8
game.update_score(User("Andrew", 8))
# prints the scoreboard

count = 0
while count < 3: 
    name = input("What is your name ")
    score = int(input("How many points? "))
    if game.get_highscore().get("score") < score:
        print(f"{name} TAKES THE LEAD")
    game.update_score(User(name, int(score)))
    count += 1

game.print_scores()

    


```
